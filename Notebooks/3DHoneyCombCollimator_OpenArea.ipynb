## X4 Alone

```python
f_x4 = '/Volumes/Storage/FOXSI/OpCal/FOXSI-3_2017Sep/X4-7Shells/CCD/rawdata/PSF/FOXSI3_X4_25mmz_CCD_10kV_0p02mA_0arcminX_0arcminY.fits'
#d_x4 = '/Volumes/Storage/FOXSI/OpCal/FOXSI-3_2017Sep/X4-7Shells/CCD/rawdata/Findingfocus/flatfield_9p56SM.fits'
d_x4c = '/Volumes/Storage/FOXSI/FOXSI-3_2018Mar/March16/PSF/Dark_FOXSI3_X4+3DC_CCD_T10Sx6_10kV_0p02mA.fits'
time_x4 = 10.

fits_x4 = pyfits.open(f_x4)
dark_x4 = pyfits.open(d_x4)

#data_x4 = (np.average(fits_x4[0].data, axis=0) - dark_x4[0].data[0])/time_x4
data_x4 = (np.average(fits_x4[0].data, axis=0) - np.average(dark_x4[0].data, axis=0))/time_x4
```

```python
fig = plt.figure(figsize=(4,4))
plt.imshow(data_x4,origin='bottom')
plt.colorbar()
```

```python
i,j = np.unravel_index(data_x4.argmax(), data_x4.shape)
w = 14
crop_data_x4 = data_x4[i-w:i+w+1,j-w:j+w+1]
fig = plt.figure(figsize=(3,3))
plt.imshow(crop_data_x4 ,origin='bottom')
plt.colorbar()
```

## X4 + 3DP Collimator

```python
f_x4c = '/Volumes/Storage/FOXSI/FOXSI-3_2018Mar/March16/PSF/FOXSI3_X4+3DC_CCD_T10Sx6_10kV_0p02mA_0arcminX_0arcminY.fits'
d_x4c = '/Volumes/Storage/FOXSI/FOXSI-3_2018Mar/March16/PSF/Dark_FOXSI3_X4+3DC_CCD_T10Sx6_10kV_0p02mA.fits'
time_x4c = 10.

fits_x4c = pyfits.open(f_x4c)
dark_x4c = pyfits.open(d_x4c)

data_x4c = (np.average(fits_x4c[0].data, axis=0) - np.average(dark_x4c[0].data, axis=0))/time_x4c
```
