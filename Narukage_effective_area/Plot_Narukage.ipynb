```python
import pandas as pd
from datetime import datetime
import csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
headers = ['E','EA','errors']
Narukage = pd.read_csv('ea_Narukage.csv',names=headers)
NE = Narukage['E']/1000
NEA = Narukage['EA']
Nerror = Narukage['errors']

X4 = pd.read_csv('EA_X4.csv',names=headers)
X4E = X4['E']
X4EA = X4['EA']
X4error = X4['errors']

# plot
plt.errorbar(NE,NEA,yerr=Nerror,marker='o',label='X5 in front of SXR detector')
plt.errorbar(X4E,X4EA,yerr=X4error,label='X4 in front of SDD')
plt.xlabel('Energy [keV]')
plt.ylabel('Effective Area [cm$^2$]')
plt.legend()
plt.show()
```
